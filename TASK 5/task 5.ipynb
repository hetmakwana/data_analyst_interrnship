```python
# Sex vs Survival
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival Count by Sex')
plt.show()
```

```python
# Histogram of Age
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30, kde=True)  # kde=True gives smooth curve
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

## Observation: Distribution of Age
Overall Shape:
The age distribution follows a right-skewed bell-shaped curve, which is common in human population data.
The majority of passengers were between 20 and 40 years old

```python
# Boxplot of Fare by Pclass
plt.figure(figsize=(8,5))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

## Observation: Fare by Passenger Class (Boxplot)
 1. Fare Distribution Across Classes:
Pclass 1 (First Class):
Has the highest median fare.
Shows a wide range of fares, including many high-value outliers (some fares exceed 500 units).
Indicates significant variability in luxury pricing within this class.

Pclass 2 (Second Class):
Has a moderate median fare, significantly lower than Pclass 1.
Narrower spread of fares compared to Pclass 1.
Few or no extreme outliers.

Pclass 3 (Third Class):
Has the lowest median fare.
Most fares are tightly packed at the lower end (below 50 units).
A few outliers are present, showing some 3rd-class passengers paid moreâ€”possibly for special accommodations.

```python
# Scatterplot Age vs Fare
plt.figure(figsize=(8,5))
sns.scatterplot(x='Age', y='Fare', data=df)
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```


##  Observation: Scatterplot â€“ Age vs Fare
1. General Distribution:
The plot is sparse and unevenly distributed, with most data points concentrated in the lower fare range (0â€“100).
Passengers of all ages paid fares in that lower range, indicating that fare was not strictly dependent on age.

2. High Fare Outliers:
A few passengers (regardless of age) paid extremely high fares (above 300), indicating they may have had premium accommodations, most likely in first class.
These high-fare outliers are scattered from young to elderly ages.

3. Dense Region:
A dense cluster appears where:
Age is between 20â€“40
Fare is between 0â€“50
This suggests most passengers were young to middle-aged adults and paid lower fares, likely indicating third-class passengers.

4. Children and Fare:
Some younger children (under 10 years) are also seen with moderate to high fares, possibly traveling with family in higher classes.
