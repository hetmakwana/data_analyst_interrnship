# SQLITE STEP

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sqlite3
```

```python
conn = sqlite3.connect('youtube.db')
df_us.to_sql('us_data', conn, if_exists='replace', index=False)
```

```python
query = """
SELECT category_id, AVG(views) as avg_views
FROM us_data
GROUP BY category_id
ORDER BY avg_views DESC;
"""
```

```python
ranked = pd.read_sql(query, conn)
```

```python
df_us['trending_date'] = pd.to_datetime(df_us['trending_date'], format='%y.%d.%m')
```

```python
df_us['days_trending'] = df_us.groupby('video_id')['trending_date'].transform(lambda x: (x.max() - x.min()).days + 1)

plt.figure(figsize=(10, 5))
sns.histplot(df_us['days_trending'], bins=20, kde=True)
plt.title("Trending Duration Distribution")
```
